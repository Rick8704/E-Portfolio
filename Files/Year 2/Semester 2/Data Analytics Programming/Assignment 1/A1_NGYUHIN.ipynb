#### QUESTION 2: Read and parse data from the```"ride.csv"```. Complete the given code (8 marks)

````python
import csv

def load_ride_data(filename):
    data = []
    with open(filename, mode='r') as file:
        reader = csv.DictReader(file) 
        #read and parse data iteratively  (5 marks)
        for row in reader:
            data.append(row)
        return data

# Write code(s) to load data from actual CSV file (2 marks)
ride_data = load_ride_data("ride.csv")

# Write code(s) to preview data (1 mark)
for record in ride_data[:3]:
    print (record)
````

#### QUESTION 3: Define a function to compute average fare by city and call the function by passing the data (10 marks)

````python
# Write your function definition here to compute average fare by city (8 marks)
def average_fare_by_city(data):
    city_fares = {}
    city_counts = {}

    for row in data:
        city = row['city']
        fare = float(row['fare'])  #Convert fare from string to float

        if city in city_fares:
            city_fares[city] += fare
            city_counts[city] += 1

        else:
            city_fares[city] = fare
            city_counts[city] = 1

    average_fares = {}
    for city in city_fares:
        average_fares[city] = city_fares[city]/city_counts[city]

    return average_fares

# Write the code(s) to call your function and pass the data (2 marks)
averages = average_fare_by_city(ride_data)
for city, avg in averages.items():
    print(f"{city}:Average Fare = ${avg:.2f}")

````

#### QUESTION 4: Define a function to summarize driver information, call the function by passing data ad display the summary (10 marks)
#### ```driver_name, num_of_rides, total_distance, total_fare```

````python

# Write your function definition here (7 marks)
def driver_info(data):
    summary = {}

    for row in data:
        driver = row['driver_name']
        distance = float(row['distance'])
        fare = float(row['fare'])
        
 # if driver not in summary
        if driver not in summary:
            summary[driver] = {
                'num_of_rides': 1,
                'total_distance': distance,
                'total_fare': fare
            }
        else:
            summary[driver]['num_of_rides'] += 1
            summary[driver]['total_distance'] += distance
            summary[driver]['total_fare'] += fare

    return summary

# Write the code(s) to call your function and pass the data (2 marks)
driver_summary = driver_info(ride_data)

# Write the code(s) to display driver's summary (1 mark)
print("Driver Summary:")
print("Driver Name  | Number of Rides | Total Distance  | Total Fare")
for driver, info in driver_summary.items():
    print(f"{driver:<12} | {info['num_of_rides']:>15} | {info['total_distance']:14.2f}  | ${info['total_fare']:.2f}")
````

#### QUESTION 5: Implement a function ```analyze_peak_slice(data, start_idx, end_idx)``` that: (10 marks)

    1.Accepts the ride data list and two indices: start_idx and end_idx.
    2.Extracts a slice of rides from start_idx = 2 to end_idx = 6 (inclusive).
    3.Calculates and returns:
        i)  Total number of rides in that slice
        ii) Average fare
        iii)List of unique drivers involved
        
    Sample output:
 
   ```Total_rides: 5```<br>
   ```Average_fare: 18.45```<br>
   ```Drivers: Ravi, Ali, Mei``` 

````python
# Write your function definition here that accepts ride data, start index and end index, slice the list, 
# calculate and returns the values in (3)   (8 marks)
def analyze_peak_slice(data, start_idx, end_idx):
    # 1. Slice the data (inclusive of end_idx, so use end_idx + 1)
    ride_slice = data[start_idx:end_idx + 1]

    # 2. Total rides in the slice
    total_rides = len(ride_slice)

    # 3. Calculate total fare
    total_fare = sum(float(ride['fare']) for ride in ride_slice)

    # 4. Calculate average fare
    average_fare = round(total_fare / total_rides, 2)

    # 5. Get list of unique drivers
    drivers = list(set(ride['driver_name'] for ride in ride_slice))

    # 6. Return result as a dictionary
    return {
        'Total_rides': total_rides,
        'Average_fare': average_fare,
        'Drivers': drivers
    }

# Write the code(s) to call your function and pass the ride data, start index and end index (2 marks)
# Call the function with index 2 to 6
result = analyze_peak_slice(ride_data, 2, 6)

# Display the result
print("Total Rides:", result['Total_rides'])
print("Average Fare:", result['Average_fare'])
print("Drivers:", ", ".join(result['Drivers']))

````

#### QUESTION 6: Use lambda function to find top 3 drivers by total fare (7 marks)

````python
# Write your function definition here for top 3 drivers by total fare (5 marks)
def get_top_3_drivers_by_fare(driver_summary):
    # Sort the driver summary dictionary by total fare using lambda
    sorted_drivers = sorted(driver_summary.items(), key=lambda x: x[1]['total_fare'], reverse=True)
    
    # Return the top 3 drivers
    return sorted_drivers[:3]

# Write the code(s) to call your function and pass the driver's summary data (2 marks)
# Call the function
top_drivers = get_top_3_drivers_by_fare(driver_summary)

# Display the results
print("Top 3 Drivers by Total Fare:")
print("Driver Name  | Total Fare")
for driver, info in top_drivers:
    print(f"{driver:<12} | ${info['total_fare']:.2f}")
````

#### QUESTION 7: Write a function to calculate average duration by cities and use lambda function to sort cities by average duration (10 marks)

````python
# Write your function definition here for average duration and sort cities (8 marks)
def sort_cities_by_avg_duration(data):
    city_summary = {}

    # Step 1: Group data by city
    for ride in data:
        city = ride['city']
        duration = float(ride['duration'])

        if city not in city_summary:
            city_summary[city] = {
                'total_duration': duration,
                'ride_count': 1
            }
        else:
            city_summary[city]['total_duration'] += duration
            city_summary[city]['ride_count'] += 1

    # Step 2: Calculate average duration
    avg_duration_by_city = []
    for city, info in city_summary.items():
        average = round(info['total_duration'] / info['ride_count'], 2)
        avg_duration_by_city.append((city, average))

    # Step 3: Sort using lambda by average duration (descending)
    sorted_cities = sorted(avg_duration_by_city, key=lambda x: x[1], reverse=True)

    return sorted_cities

# Write the code(s) to call your function and pass the data (2 marks)
# Call the function
sorted_city_durations = sort_cities_by_avg_duration(ride_data)

# Display result
print("Cities Sorted by Average Duration:")
print("City         | Average Duration")
for city, avg_duration in sorted_city_durations:
    print(f"{city:<12} | {avg_duration:.2f}")
````

#### QUESTION 8: Define a function to write driver summary to a text file ```driver_summary.txt``` with the following information:
#### ```"Driver", "Total Rides", "Total Distance", "Total Fare"```

````python
# Write your function definition here for writing driver summary into a text file (8 marks)
def write_driver_summary_to_file(driver_summary, filename):
    with open(filename, mode='w') as file:
        # Write the header
        file.write("Driver       | Total Rides | Total Distance | Total Fare\n")
        file.write("----------------------------------------------------------\n")

        # Write each driver's info
        for driver, info in driver_summary.items():
            file.write(f"{driver:<12} | {info['num_of_rides']:>11} | {info['total_distance']:>14.2f} | ${info['total_fare']:.2f}\n")

# Write the code(s) to call your function and pass the driver summary data (2 marks)
# Call the function and pass the driver summary
write_driver_summary_to_file(driver_summary, "driver_summary.txt")
````
